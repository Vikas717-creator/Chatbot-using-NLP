```python
import json
with open('chatbot.json', 'r') as file:
    intents_data = json.load(file)

# Extract examples and labels
examples = []
labels = []

for intent in intents_data['intents']:
    for example in intent['examples']:
        examples.append(example)
        labels.append(intent['intent'])
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab') # Download the missing 'punkt_tab' resource

# Preprocessing function
def preprocess_text(text):
    lemmatizer = WordNetLemmatizer()
    stop_words = set(stopwords.words('english'))
    tokens = word_tokenize(text.lower())  # Tokenize and lowercase
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word.isalpha() and word not in stop_words]
    return ' '.join(tokens)

# Apply preprocessing
examples = [preprocess_text(example) for example in examples]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Convert text to TF-IDF features
tfidf = TfidfVectorizer(max_features=10000, ngram_range=(1, 3))
X = tfidf.fit_transform(examples).toarray()
```

```python
import json
import random
import nltk
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
from nltk.corpus import stopwords
import string
import joblib

# Download NLTK resources
nltk.download('stopwords')
nltk.download('punkt')

# Preprocess text
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    tokens = nltk.word_tokenize(text.lower())
    tokens = [t for t in tokens if t not in string.punctuation and t not in stop_words]
    return " ".join(tokens)

class Chatbot:
    def __init__(self, dataset_path=None, model_path=None):
        self.vectorizer = None
        self.classifier = None
        self.responses = {}
        if model_path:
            self._load_model(model_path)
        elif dataset_path:
            self.dataset_path = dataset_path
            self._load_dataset()
            self._train()
        else:
            raise ValueError("Either dataset_path or model_path must be provided.")

    def _load_dataset(self):
        with open(self.dataset_path, 'r') as file:
            data = json.load(file)
        self.examples = []
        self.labels = []
        for intent_data in data["intents"]:
            intent = intent_data["intent"]
            self.examples.extend([preprocess_text(example) for example in intent_data["examples"]])
            self.labels.extend([intent] * len(intent_data["examples"]))
            self.responses[intent] = intent_data["responses"]

    def _train(self):
        self.vectorizer = CountVectorizer()
        X = self.vectorizer.fit_transform(self.examples)
        y = self.labels
        self.classifier = RandomForestClassifier()
        self.classifier.fit(X, y)

    def save_model(self, model_path):
        model_data = {
            "vectorizer": self.vectorizer,
            "classifier": self.classifier,
            "responses": self.responses
        }
        joblib.dump(model_data, model_path)  # Save as .joblib
        print(f"Model saved to {model_path}")

    def _load_model(self, model_path):
        model_data = joblib.load(model_path)  # Load the model from .joblib
        self.vectorizer = model_data["vectorizer"]
        self.classifier = model_data["classifier"]
        self.responses = model_data["responses"]
        print(f"Model loaded from {model_path}")

    def get_response(self, user_input):
        processed_input = preprocess_text(user_input)
        vectorized_input = self.vectorizer.transform([processed_input])
        try:
            predicted_intent = self.classifier.predict(vectorized_input)[0]
            return random.choice(self.responses[predicted_intent])
        except:
            return "Sorry, I didn't understand that."

# Example usage (for training and saving the model)
if __name__ == "__main__":
    chatbot = Chatbot("chatbot1.json")
    chatbot.save_model("chatbot_model.joblib")  # Save as .joblib
```

```python
!pip install sentence-transformers nltk scikit-learn imbalanced-learn
import joblib
import json
import nltk
import numpy as np
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sentence_transformers import SentenceTransformer
from sklearn.feature_extraction.text import TfidfVectorizer

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')

# Load intents data
with open('chatbot1.json', 'r') as file:
    intents_data = json.load(file)

# Preprocessing function (experiment with options)
def preprocess_text(text, option='lemmatize'):
    tokens = word_tokenize(text.lower())
    if option == 'lemmatize':
        lemmatizer = WordNetLemmatizer()
        tokens = [lemmatizer.lemmatize(word) for word in tokens if word.isalpha()]
    elif option == 'stopwords':
        stop_words = set(stopwords.words('english'))
        tokens = [word for word in tokens if word.isalpha() and word not in stop_words]
    return ' '.join(tokens)

# Extract examples and labels, keeping track of original indices
examples = []
labels = []


for intent in intents_data['intents']:
    for example in intent['examples']:
        examples.append(preprocess_text(example, option='lemmatize'))
        examples.append(preprocess_text(example, option='stopwords'))
        examples.append(example)
        labels.extend([intent['intent']] * 3)

# 1. TF-IDF Feature Extraction
tfidf_vectorizer = TfidfVectorizer(max_features=1000, ngram_range=(1, 2))
X_tfidf = tfidf_vectorizer.fit_transform(examples).toarray()

# 2. Sentence-BERT Embeddings
model_sbert = SentenceTransformer('all-mpnet-base-v2')
X_sbert = model_sbert.encode(examples)

# 3. Combine Features
X = np.concatenate([X_sbert, X_tfidf], axis=1)

# Encode labels
encoder = LabelEncoder()
y = encoder.fit_transform(labels)

# Split data with StratifiedKFold, using original X and y
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
for train_index, test_index in skf.split(X, y):  # Use original X and y
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    # Train Random Forest with Hyperparameter Tuning
    param_grid = {
        'n_estimators': [50, 100, 200],
        'max_depth': [None, 10, 20],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4]
    }

    grid_search = GridSearchCV(RandomForestClassifier(random_state=42),
                               param_grid, cv=5, scoring='accuracy', n_jobs=-1)
    grid_search.fit(X_train, y_train)

    best_rf_model = grid_search.best_estimator_
    y_pred = best_rf_model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print("Random Forest Accuracy (with Sentence-BERT, TF-IDF):", accuracy)

# Save the trained model using joblib
joblib.dump(best_rf_model, 'chatbot1.joblib')
```
